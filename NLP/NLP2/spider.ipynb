```python
# -*- coding: utf-8 -*-
# @Time : 2022/12/10 17:28
# @Author : menike
# @File : nlp_spider.py
# @Software: PyCharm
import requests
import json
import csv
import time

f = open(
        f'./data/nlp_spider.csv',
        mode='a',
        encoding='utf-8-sig',
        newline='')
# 创建一个csv文件，mode=a表示对文件只能写入，encoding是内容文字，newline避免有换行字符等产生
csv__ = csv.DictWriter(
    f,
    fieldnames=(
        '评论发表时间',
        '评论内容',

    )
)
# f是创建的csv文件，fieldnames表示列名
csv__.writeheader()
for page in range(2,100):
    print(f"当前爬取的是第{page}页评论")
    url = f"https://m.weibo.cn/api/container/getIndex?containerid=231522type%3D1%26t%3D10%26q%3D%23%E7%A5%9E%E8%88%9F%E5%8D%81%E4%BA%94%23&extparam=%23%E7%A5%9E%E8%88%9F%E5%8D%81%E4%BA%94%23&luicode=10000011&lfid=231522type%3D1%26t%3D10%26q%3D%23%E7%A5%9E%E5%8D%81%E4%BA%94%E5%B7%B2%E5%81%9A%E5%A5%BD%E5%87%BA%E5%BE%81%E5%87%86%E5%A4%87%23&page_type=searchall&page={page}"
    headers = {
        'Accept': 'application/json,text/plain,*/*',
        'Referer': "https://m.weibo.cn/search?containerid=231522type%3D1%26t%3D10%26q%3D%23%E7%A5%9E%E8%88%9F%E5%8D%81%E4%BA%94%23&extparam=%23%E7%A5%9E%E8%88%9F%E5%8D%81%E4%BA%94%23&luicode=10000011&lfid=231522type%3D1%26t%3D10%26q%3D%23%E7%A5%9E%E5%8D%81%E4%BA%94%E5%B7%B2%E5%81%9A%E5%A5%BD%E5%87%BA%E5%BE%81%E5%87%86%E5%A4%87%23",
        'User-Agent': 'Mozilla/5.0(WindowsNT10.0;Win64;x64)AppleWebKit/537.36(KHTML,likeGecko)Chrome/108.0.0.0Safari/537.36Edg/108.0.1462.46',
        'X-Requested-With': 'XMLHttpRequest',


    }
    try:
        r = requests.get(url, headers=headers, timeout=31)
    except requests.HTTPError as e:
        print(e)
        print("HTTPError")
    except requests.RequestException as e:
        print(e)
    except:
        print("Unknown Error !")
    s = json.loads(r.text)
    for i in range(9):
        comment_time = s['data']['cards'][i]["card_group"][0]["mblog"]["created_at"]
        # fans = s['data']['cards'][i]["card_group"][0]["mblog"]["fans"]
        # status_city = s['data']['cards'][i]["card_group"][0]["mblog"]["status_city"]
        # status_country = s['data']['cards'][i]["card_group"][0]["mblog"]["status_country"]
        text = s['data']['cards'][i]["card_group"][0]["mblog"]["text"]
        print(text)
        print()
        dit = {
            '评论发表时间': comment_time,
            '评论内容': text,
        }
        # 把拆分的数据整合进一个新的字典
        csv__.writerow(dit)
    time.sleep(3)
```
